```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
from html.parser import HTMLParser
import re
```

```python
data = requests.get('https://www.basketball-reference.com/leagues/NBA_stats_per_game.html')
soup = BeautifulSoup(data.text, 'html.parser')
div = soup.find('div' , {'class': 'overthrow table_container', 'id': 'div_stats'})
table = div.find('table')
tbody = table.find('tbody')
html = table.encode(encoding='ISO-8859-1')
dfs = pd.read_html(html,header=0)
df = dfs[0]
df.to_csv('all_seasons.csv', encoding='ISO-8859-1')
df
```
